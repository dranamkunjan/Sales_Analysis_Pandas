# Step1:Clean the data

```python
nan_df=df[df.isna().any(axis=1)]
```

```python
nan_df.head()
```

```python
df=df.dropna(how='all')  #above it is seen rows with nan values, so we can drop all
```

# Step2:Add month column

```python
df['Month']= df['Order Date'].str[0:2]  #we use index to extract 04 for month column
df
```

```python
#next is to convert above month to int
df['Month']= df['Order Date'].str[0:2].astype('int32')
```

#ValueError: invalid literal for int() with base 10: 'Or'
#next step would be to filter and check Month column by Or.

```python
temp_df = df[df['Month'].str[0:2]=='Or']
```

```python
temp_df  #355 rows are observed for Month Or, next step would be to filter this out of df

#we filter this out in original df
```

```python
df = df[df['Order Date'].str[0:2]!='Or']  #duplicated coulmn headers 'Or'
```

```python
#now we can convert Month column to int

df['Month']= df['Order Date'].str[0:2].astype('int32')
```

```python
df
```

# Step3:We would want all columns to correct type, since we will perform mathematical operation for total sales

```python
df=['Quantity Ordered'] #make int
df=['Price Each']  #make float
```

```python
df['Quantity Ordered']=pd.to_numeric(df['Quantity Ordered'])
df['Price Each'] =pd.to_numeric(df['Price Each']) 
df.head()
```

# step4:add another (column)-sales column

```python
df['Sales']=df['Quantity Ordered'] *df['Price Each']
```

```python
df
```

```python
#analysis:best month for sales and amount for sales
```

```python
df_grouped = df.groupby('Month').sum()
```

```python
df_month_sales = df_grouped[['Sales', 'Quantity Ordered','Price Each']]
```

```python
df_month_sales
```

# step4:Quickly plot the results in graph

```python
import matplotlib.pyplot as plt
months = range(1,13)
plt.bar(months,df_month_sales['Sales'])
plt.xticks(months)
plt.xlabel('Month number')
plt.ylabel('Sales in USD($)')
plt.show
```
